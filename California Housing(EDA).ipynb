# California Housing Project

### The data pertains to the houses found in a given California district and some summary stats about them based on the 1990 census data. It contains one row per census block group. A block group is the smallest geographical unit for which the U.S. Census Bureau publishes sample data (a block group typically has a population of 600 to 3,000 people).
### An household is a group of people residing within a home. Since the average number of rooms and bedrooms in this dataset are provided per household, these columns may take surpinsingly large values for block groups with few households and many empty houses, such as vacation resorts.
### The data aren't cleaned so there are some preprocessing steps required.
 
### -Data has 10 coloumns with 20,640 observations.
### -Data belongs to the state of California located in USA.
### -Data also describes the exact location of the houses, demography of the inmates, incomes,their property values and their proximity from the ocean

### Columns and their datatypes:

### 1. longitude(Quantitative/Numerical/Continuous): A measure of how far west a house is; a higher value is farther west

### 2. latitude(Quantitative/Numerical/Continuous): A measure of how far north a house is; a higher value is farther north

### 3. housing_median_age(Quantitative/Numerical/Continuous): Median age of a house within a block; a lower number is a newer building

### 4. total_rooms(Quantitative/Numerical/Discrete): Total number of rooms within a block

### 5. total_bedrooms(Quantitative/Numerical/Discrete): Total number of bedrooms within a block

### 6. population(Quantitative/Numerical/Discrete): Total number of people residing within a block

### 7. households(Quantitative/Numerical/Discrete): Total number of households, a group of people residing within a home unit, for a block

### 8. median_income(Quantitative/Numerical/Continuous): Median income for households within a block of houses (measured in tens of thousands of US Dollars)

### 9. median_house_value(Quantitative/Numerical/Continuous): Median house value for households within a block (measured in US Dollars)

### 10. ocean_proximity(Qualitative/Categorical/Ordinal): Location of the house w.r.t ocean/sea dependent on latitude and longitude

### Goals:
### The given dataset can be used to predict the housing prices in the future and also the various factors affecting it.

### 10) Please create a new column named total_bedroom_size. If size 10 or less. It should be quoted as small. If the size is 11 or more but 1000 less, it should be medium, otherwise, it should be large.

```python
conditions=[
    (df_housing["total_bedrooms"]<=10),
    (df_housing["total_bedrooms"]<=1000),
    (df_housing["total_bedrooms"]>1000)
] #Conditions for the new column
values=["Small","Medium","Large"] #Appropriate corresponding to the above conditions for the new column
df_housing["total_bedroom_size"]= np.select(conditions,values) #Creating the new column with 'select' function
df_housing
```

#### The above dataframe has a new column called 'total_bedroom_size' with required conditions.
