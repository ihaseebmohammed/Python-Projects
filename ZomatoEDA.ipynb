```python
df_cc= pd.read_excel("/content/Country-Code[1].xlsx")
df_cc.head()
```

```python
df_cc.columns
```

```python
df.columns
```

```python
final_df= pd.merge(df,df_cc, on = 'Country Code',how="left")
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
#pie-chart
plt.pie(x=country_values[:3],labels = country_names[:3],autopct="%1.2f%%")
```

###Observation: 
###a)The majority of records from the data set are from India
###b)Followed by USA and UK

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

###Observations:
####a) The dataset contains large number customers who haven't rated the restaurant(2148)
####b) The rating between 4.5 to 4.9 indicate the customers have really liked the food and spared their time in rating it(Excellent)
####c) The rating 4 to 4.4 have very good rating
####d) The rating  3.4 to 3.9 -Good rating
####e) 2.5 to 3.4 -Average
####f) Below 2.5 is poor




```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue ="Rating color",data=ratings,palette =["White","Red","Orange","Yellow","Green", "Green"])
```

###Observations:
a)Not rated count is very high
b)Maximum ratings lie between 2.5 and 3.4

```python
sns.countplot(x="Rating color",data=ratings,palette =["White","Red","Orange","Yellow","Green", "Green"])
sns.set_style('darkgrid')
```

```python
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index().rename(columns={0:"Zero Ratings"})
```

###Maximum 0 ratings are from Indian customers

```python
final_df.groupby(["Country","Currency"]).size().reset_index().rename(columns={0:"size"}).sort_values(by="size",ascending=False)
```

```python
final_df[final_df["Has Online delivery"]=='Yes'].groupby("Country").size().reset_index().rename(columns={0:"Has delivery service"})
```
