####Dataset shows the sortable list of countries by their population in the year **2022**. There are **235** countries along with their population. And there are **11** columns each representing different features of countries. This list includes both countries and dependent territories. ####In demographics, the term world population is often used to refer to the total number of humans currently living, and was estimated to have exceeded 7.9 billion as of November 2021. It took over **two million** years of human prehistory and history for the human population to reach **one billion** and only **200** years more to grow to **7 billion**. Worldometers, is a reference website that provides counters and real-time statistics for diverse topics

###**Columns and their datatypes:**
####1)**Country (or dependency**)-(Qualtitaive/Categorical/Nominal):This column contains different country's name (235 countries).
####2)**Population (2020)**-(Quantitative/Numerical/Discrete): This columns contains the population of different countries.
####3)**Yearly Change**-(Quantitative/Numerical/Continuous): This columns contains the population change by year in terms of percentage.
####4)**Net Change	Density**-(Quantitative/Numerical/Continuous): It is calculated by dividing the number of people added to a population in a year.
####5)**Density (P/Km)**-(Quantitative/Numerical/Continuous): To calculate the population density, you will divide the population by the size of the area.
####6)**Land Area (Km)**-(Quantitative/Numerical/Continuous): Land area is the area in square kilometres of the land-based portions of standard geographic areas
####7)**Migrants (net)**-(Quantitative/Numerical/Discrete):Net migration is the difference between the number of people moving into an area (a country, state, or county, for example) and the number moving out.
####8)**Fert. Rate**-(Quantitative/Numerical/Continuous):The total fertility rate in a specific year is defined as the total number of children that would be born to each woman if she were to live to the end of her child-bearing years and give birth to children in alignment with the prevailing age-specific fertility rates.
####9)**Med. Age**-(Quantitative/Numerical/Continuous): Median age is the age that divides a population into two numerically equally sized groups; that is, half the people are younger than this age and half are older.
####10)**Urban Pop %**-(Quantitative/Numerical/Continuous): Urban population refers to the population inhabiting areas that have a greater population density than rural areas and are overall more compact than rural areas
####11)**World Share**- Percentage contribution of each country to the world population.

####**Goals**: To predict and analyse the population change in the coming years and of the past and also various features associated with it.

```python
#Importing required libraries for the webscraping and analysis
import pandas as pd #for data analysis, data manipulation, and cleaning of data
from bs4 import BeautifulSoup #Beautiful Soup is a Python package for parsing HTML data
import requests #The requests module allows you to send HTTP requests using Python
```

```python
URL= "https://www.worldometers.info/world-population/population-by-country/" #Copying the required URL for parsing

page = requests.get(URL) #Make a request to the URL to get the HTML of the page

soup = BeautifulSoup(page.content,"html.parser") #Extracting the data

print(soup.prettify()) #method will turn a Beautiful Soup parse tree into a nicely formatted Unicode string
```

```python
get_title= soup.title #To get the title of the table required
print(get_title.text)
```

```python
table = soup.find_all('table',{"id":"example2"}) #Extracting table for the analysis
```

```python
data_frame= pd.read_html(str(table))[0] #Reading the table as pandas dataframe
```

```python
data_frame #Required data frame
```

###* **Create appropriate graphs to examine the distribution of Population, median age and Urban population and Yearly change.**

```python
#Importing required libraries for the analysis
import numpy as np #for mathematical functions
import matplotlib.pyplot as plt #for plotting numerical data
import seaborn as sns #for data visualization
import matplotlib.ticker as mtick #for 2D plots
import matplotlib.colors as mcolors #for more colours in the plots
```

####**Population Distribution**

```python
ax= sns.distplot(data_frame['Population (2020)'], hist=True, kde=False, color = 'magenta', bins=int(180/2),
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4}) #Plotting distplot to know the distribution of Tenure Months column
sns.set_style("darkgrid")
ax.set_ylabel('Country') #label x axis
ax.set_xlabel('Population') #label y axis
ax.set_title('Population Distribution') #Assigning the title to the chart
```

```python
plt.figure(figsize=(10,5))   #defining the figure size

font1 = {'family':'serif','color':'red','size':20}   #Giving font styles for graph
font2 = {'family':'serif','color':'blue','size':15} #Giving font styles for graph

sns.distplot(data_frame['Population (2020)'],color = 'RED')  #Distribution plot to see rating with blue color

plt.xlabel('Country',fontdict = font2)
plt.ylabel('Density',fontdict = font2)
plt.title('Distribution of Population',fontdict = font1)  # printing title
```
